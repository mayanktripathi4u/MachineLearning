# Date Range - Pandas date_range function

Time series analysis is crucial in financial data analysis space. Pandas has in built support of time series functionality that makes analyzing time series extremely efficient. Here, we will see how date_range function allows to generate datetimeindex with specific start and end dates. It can also generate periods with different frequencies such as hourly,daily,monthly, weekly etc. We will then cover how asfreq function can be used to resample dataframe to a different frequency.

https://github.com/codebasics/py/blob/master/pandas/15_ts_date_range/aapl_no_dates.csv

```python
import pandas as pd
```

```python
df = pd.read_csv("aapl_no_dates.csv")
df.head()
```

```python
rng = pd.date_range(start = "6/1/2017", end = '6/30/2017', freq = 'B')
rng
```

```python
df.set_index(rng, inplace=True)
df
```

```python
%matplotlib inline
df.Close.plot()
```

```python
df["2017-06-01" : "2017-06-10"]
```

```python
# How to include week-end. Assuming on week-end the stock price will remain same as of Friday.
df.asfreq('D', method = 'pad')
# For this we have used pad method, which will add the same data of last working day to week-end.
```

```python
rng_prd = pd.date_range(start = '1/1/2017', periods = 72, freq = 'B')
rng_prd
# This will generate the date range starting from 01 Jan 2017 to 72 periods as we do not have defined end date.
```

```python
# Generating sample data-set
import numpy as np
np.random.randint(1,10, len(rng_prd))
```

```python
ts = pd.Series(np.random.randint(1,10, len(rng_prd)), index = rng_prd)
ts.head()
```
